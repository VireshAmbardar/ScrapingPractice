```python


## itterate upto last page

counter =1
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}

last_page = 933
#total no of pages are 933
for i in range(1,last_page+1):
    url = f"https://www.shutterstock.com/search/squat?page={i}"
    webpage = requests.get(url,headers=headers).text

    soup = BeautifulSoup(webpage ,"html.parser")
    
    for i in (soup.find_all("div", class_="mui-1nl4cpc-gridContainer-root")):
        img_tag = soup.find_all('img')
        for j in tqdm(img_tag):
            img_url  = j['src']
            try:
                response = requests.get(img_url)
            
                open(f"image_{counter}.jpg", "wb").write(response.content)
                counter+=1
            except:
                pass
```
