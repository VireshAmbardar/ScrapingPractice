```python
 import requests

url = 'https://quotes.toscrape.com/page/1/'
res = requests.get(url)
print(res.text , res.encoding)
```

```python
from bs4 import BeautifulSoup as bs

soup = bs(res.text, 'html.parser')
soup
```

```python
print(bs.prettify(soup))
```

```python
print(soup.title)
```

```python
print(soup.title.string)
```

```python
print(soup.span)
```

```python
print(soup.title.parent)
```

```python
print(soup.title.parent.name)
```

```python
# <span class="text" itemprop="text">
print(soup.find_all('span'))  # pass anchor tag(starts with <'spam'>)  
```

```python
# to get quote we have to give class also. so

#<span class="text" itemprop="text">“A day without sunshine is like, you know, night.”</span>, <span>by <small class

for tag in soup.find_all('span' ,{'class' : 'text',   #we can pas clas and various clases to specify according to our need
                                 'itemprop':"text"}):
    
    print(tag.string)
```

```python
# to get authors
# <small class="author" itemprop="author">
#         Albert Einstein
#        </small>

for tag in soup.find_all('small' ,{'class':"author"}):
    print(tag.string)
```

```python
import pandas as pd
```

```python
# lets make a txt file here and save authors and quotes
author=[]
quote=[]

with open('bs4-arthur-quotes','w') as f:
    for i in range(1,6):
        res = requests.get('https://quotes.toscrape.com/page/{}'.format(i))
        soup = bs(res.text, 'html.parser')
        
        for tag1 in soup.find_all('small' ,{'class':"author"}):
            print(tag1.string)
            author.append(tag1.string)
            
            
        for tag in soup.find_all('span' ,{'class' : 'text',  
                                         'itemprop':"text"}):
            print(tag.string)
            quote.append(tag.string)
            
quotes_df = pd.DataFrame({'author':author , 'quotes':quote})
quotes_df.head()
```
