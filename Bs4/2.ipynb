```python
## get top 250 movies 

import requests

res = requests.get('https://www.imdb.com/chart/top/')
res.status_code
```

```python
# so we are good to go
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text,'html.parser')
```

```python
soup
```

```python
tbody = soup.find('tbody',{"class" : "lister-list"})
tbody
```

```python
# under atble body we have various table rows
#we have to enter in each table row
trs = tbody.find_all('tr')
trs
```

```python
trs[0]
```

```python
#so we are going to iterating in each tr bcz tqabrows is a list so  hance find

for tabrow in trs:
    tab_data = tabrow.find('td',{'class':'titleColumn'}) # td means table data- each list ahs 1 same attribute
    taba = tab_data.a
    print(taba.attrs)   # here we can get director name and lead name of movie
    

# so this is how we will extract data from a website
```

```python
##to get the movie name which is tring in a tag is
for tabrow in trs:
    tab_data = tabrow.find('td',{'class':'titleColumn'}) 
    taba = tab_data.a
    print(taba.string)
```
